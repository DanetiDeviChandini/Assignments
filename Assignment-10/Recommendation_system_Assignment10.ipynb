```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("book.csv",encoding='latin1')
dataset
```

```python
dataset2 = dataset.iloc[:,1:]
dataset2
```

```python
#number of unique users in the dataset
len(dataset2['User.ID'].unique())
```

```python
#number of books in data set
len(dataset2['Book.Title'].unique())
```

```python
##user item intraction matrix
# pivot table summerising data for data ananlysis
dataset3=dataset.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
dataset3
```

```python
#index values replacing with unique values
dataset3.index = dataset2['User.ID'].unique()
dataset3
```

```python
#Impute those NaNs with 0 values
dataset3.fillna(0, inplace=True)
```

```python
dataset3
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( dataset3.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim1 =pd.DataFrame(user_sim)
user_sim1
```

```python
#Set the index and column names to user ids 
user_sim1.index=dataset2['User.ID'].unique()
user_sim1.columns=dataset2['User.ID'].unique()
user_sim1
```

```python
# fill diagnol value with 0
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim1
```

```python
#Most Similar Users, idmax show column with highest cosine similarity values
user_sim1.idxmax(axis=1)[0:]
```
